```python
import requests
import json
tools=[
    {
        "type":"function",
        "name":"get_departure_arrival_codes",
        "description":"Understand deeply the user input and get a particular code for departure and arrival locations as registered in airports",
        "parameters":{
            "type":"object",
            "properties":
            {
                "departure_code":{
                    "type":"string",
                    "description":"Get a specific code for departure location as registered in airports"
                },
                "arrival_code":{
                    "type":"string",
                    "description":"Get a specific code for arrival location as registered in airports"
                }
            },
            "required":["departure_code","arrival_code"],
        },
    },
]
def getting_flights(departure,arrival):
    base_URL="https://api.aviationstack.com/v1/flights"
    api="Paste_Your_Avitation_ApiKey"
    params={
    "access_key":api,
    "dep_iata":departure,
    "arr_iata":arrival,
    }
    flight_res=requests.get(base_URL,params=params)
    result=flight_res.json()
    final_result=result['data'][0]
    final={
        "Flight date":final_result["flight_date"],
        "Departure":final_result["departure"]["airport"],
        "Departure iata":final_result["departure"]["iata"],
        "Arrival":final_result["arrival"]["airport"],
        "Arrival iata":final_result["arrival"]["iata"],
        "Airline":final_result["airline"]["name"],
        "Flight number":final_result["flight"]["number"],
        "Flight iata":final_result["flight"]["iata"]
    }
    return(final)
print("Hey I am your flight assistant!")
content=input("How can I help you  ")
input_list=[
    {
        "role":"user","content":content
    }
]
response=client.responses.create(
    model="gpt-4o-mini",
    tools=tools,
    input=input_list,
)

input_list+=response.output
for item in response.output:
    if item.type=="function_call":
        if item.name=="get_departure_arrival_codes":
            flights=getting_flights(json.loads(item.arguments)["departure_code"],json.loads(item.arguments)["arrival_code"])
            input_list.append({
                "type":"function_call_output",
                "call_id":item.call_id,
                "output":json.dumps({
                    "flights":json.dumps(flights)
                })
            })
    #print(json.dumps(flights))
fina=client.responses.create(
    model="gpt-4o-mini",
    instructions="Respond human-friendly with the flight details. Give in a paragraph",
    tools=tools,
    input=input_list,
)
fina.output[0].content[0].text
```
